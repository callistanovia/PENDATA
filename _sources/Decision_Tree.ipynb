# **Decision Tree**

## **Definisi Decision Tree:**

**Decision Tree atau Pohon Keputusan** adalah salah satu metode dalam teknik data mining dan machine learning yang digunakan untuk membuat keputusan atau prediksi berdasarkan data yang ada. Decision Tree menyajikan model prediksi berbentuk struktur pohon, di mana setiap simpul (node) mewakili suatu atribut, setiap cabang mewakili suatu kondisi/hasil dari atribut tersebut, dan setiap daun (leaf node) mewakili hasil akhir atau keputusan.

###  **3. Ambil contoh threshold split (misalnya):**

* Threshold = 90 mm

 Maka kita akan membagi:

* Group Left (≤ 90 mm)
* Group Right (> 90 mm)

###  **4. Hitung Entropy Awal (sebelum split)**

$$
Entropy(D) = -p_1 \log_2(p_1) - p_2 \log_2(p_2)
$$

$$
p_1 = \frac{8}{15},\quad p_2 = \frac{7}{15}
$$

$$
Entropy(D) = -\frac{8}{15} \log_2 \frac{8}{15} - \frac{7}{15} \log_2 \frac{7}{15} \approx 0.996
$$


###  **5. Entropy setelah split**

Group Left (≤ 90 mm):

* Data: No 13, 9, 6, 15, 8, 7, 4, 10 → 8 data
* Kelas: 7 Mangifera odorata, 1 Mangifera indica

$$
Entropy(left) = -\frac{7}{8}\log_2\frac{7}{8} - \frac{1}{8}\log_2\frac{1}{8} \approx 0.543
$$

Group Right (> 90 mm):

* Data: No 12, 3, 1, 14, 2, 11, 5 → 7 data
* Kelas: 7 Mangifera indica

$$
Entropy(right) = -\frac{7}{7} \log_2 \frac{7}{7} = 0
$$

Information Gain:

$$
IG = Entropy(D) - \left( \frac{8}{15} \cdot 0.543 + \frac{7}{15} \cdot 0 \right) \approx 0.996 - 0.2896 \approx \boxed{0.7064}
$$

###  **6. Gini Index**

$$
Gini(D) = 1 - \sum p_i^2 = 1 - \left(\frac{8}{15}\right)^2 - \left(\frac{7}{15}\right)^2 \approx 0.497
$$

Group Left (≤ 90 mm):

* 7 Mangifera odorata, 1 Mangifera indica

$$
Gini(left) = 1 - \left(\frac{7}{8}\right)^2 - \left(\frac{1}{8}\right)^2 = 1 - 0.765625 - 0.015625 = 0.21875
$$

Group Right (> 90 mm):

* 7 Mangifera indica

$$
Gini(left) = 1 - \left(\frac{7}{7}\right)^2 = 1 - 1 = 0
$$

Gini Split:

$$
Gini_{split} = \frac{8}{15} \cdot 0.21875 + \frac{7}{15} \cdot 0 = \boxed{0.116}
$$

###  **3. Komposisi kelas:**

* 7 *Mangifera odorata*

###  **4. Entropy setelah split:**

#### Group Left (≤ 280 gram):

Data: No 13, 9, 6, 15, 8, 7, 4 → 7 data
Kelas: 6 Mangifera odorata, 1 Mangifera indica

$$
Entropy(left) = -\frac{6}{7}\log_2\frac{6}{7} - \frac{1}{7}\log_2\frac{1}{7}
$$

$$
= -0.857 \times \log_2(0.857) - 0.143 \times \log_2(0.143) \approx 0.591
$$

#### Group Right (> 280 kg):

* Data: No 10, 1, 3, 14, 2, 5, 12, 11 → 8 data
* Kelas: 7 Mangifera indica, 1 Mangifera odorata

$$
Entropy(right) = -\frac{7}{8}\log_2\frac{7}{8} - \frac{1}{8}\log_2\frac{1}{8}
$$

$$
= -0.875 \times \log_2(0.875) - 0.125 \times \log_2(0.125) \approx 0.543
$$


###  **5. Information Gain**

$$
IG = Entropy(D) - \left( \frac{7}{15} \cdot 0.591 + \frac{8}{15} \cdot 0.543 \right)
$$

$$
IG = 0.996 - (0.276 + 0.29) = 0.996 - 0.566 = \boxed{0.43}
$$


###  **6. Gini Index**

$$
Gini(D) = 1 - \left(\frac{8}{15}\right)^2 - \left(\frac{7}{15}\right)^2 = 1 - 0.284 - 0.218 = 0.498
$$

#### Gini(Left):

$$
Gini(left) = 1 - \left(\frac{6}{7}\right)^2 - \left(\frac{1}{7}\right)^2 = 1 - 0.7347 - 0.0204 = 0.2449
$$

#### Gini(Right):

$$
Gini(right) = 1 - \left(\frac{7}{8}\right)^2 - \left(\frac{1}{8}\right)^2 = 1 - 0.7656 - 0.0156 = 0.2188
$$

#### Gini Split:

$$
Gini_{split} = \frac{7}{15} \cdot 0.2449 + \frac{8}{15} \cdot 0.2188 = 0.114 + 0.117 = \boxed{0.231}
$$

## **Visualisasi Decision Tree**

Ini adalah code program untuk mencari decision tree sekaligus rules

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_text
from sklearn.preprocessing import LabelEncoder

# === Data Buah Mangifera ===
data = {
    "Diameter Buah (mm)": [95, 100, 92, 90, 98, 80, 85, 82, 78, 88, 105, 96, 75, 95, 80],
    "Berat Buah (gram)": [300, 320, 310, 280, 330, 250, 270, 260, 240, 290, 350, 340, 230, 310, 260],
    "Kelas": [
        "Mangifera indica", "Mangifera indica", "Mangifera indica", "Mangifera indica", "Mangifera indica",
        "Mangifera odorata", "Mangifera odorata", "Mangifera odorata", "Mangifera odorata", "Mangifera odorata",
        "Mangifera indica", "Mangifera indica", "Mangifera odorata", "Mangifera indica", "Mangifera odorata"
    ]
}
df = pd.DataFrame(data)

# === Label Encoding ===
le = LabelEncoder()
df['Label'] = le.fit_transform(df['Kelas'])  # Mangifera indica = 1, Mangifera odorata = 0 (atau sebaliknya)

# === Fitur dan Target ===
X = df[["Diameter Buah (mm)", "Berat Buah (gram)"]]
y = df["Label"]

# === Train Decision Tree ===
clf = DecisionTreeClassifier(criterion="gini", max_depth=3, random_state=42)
clf.fit(X, y)

# === 1. Tampilkan Pohon Keputusan ===
print("\n=== Pohon Keputusan (Decision Tree) ===")
tree_rules = export_text(clf, feature_names=list(X.columns))
print(tree_rules)
```

## **Visualisasi Splitting**

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_text
from sklearn.preprocessing import LabelEncoder

# === Data Buah ===
data = {
    "Diameter Buah (mm)": [95, 100, 92, 90, 98, 80, 85, 82, 78, 88, 105, 96, 75, 95, 80],
    "Berat Buah (gram)": [300, 320, 310, 280, 330, 250, 270, 260, 240, 290, 350, 340, 230, 310, 260],
    "Kelas": [
        "Mangifera indica", "Mangifera indica", "Mangifera indica", "Mangifera indica", "Mangifera indica",
        "Mangifera odorata", "Mangifera odorata", "Mangifera odorata", "Mangifera odorata", "Mangifera odorata",
        "Mangifera indica", "Mangifera indica", "Mangifera odorata", "Mangifera indica", "Mangifera odorata"
    ]
}
df = pd.DataFrame(data)

# === Label Encoding ===
le = LabelEncoder()
df['Label'] = le.fit_transform(df['Kelas'])  # Mangifera indica = 1, odorata = 0 (atau sebaliknya)

# === Fitur dan Target ===
X = df[["Diameter Buah (mm)", "Berat Buah (gram)"]]
y = df["Label"]

# === Train Decision Tree ===
clf = DecisionTreeClassifier(criterion="gini", max_depth=3, random_state=42)
clf.fit(X, y)

# === 1. Tampilkan Pohon Keputusan ===
print("\n=== Pohon Keputusan (Decision Tree) ===")
tree_rules = export_text(clf, feature_names=list(X.columns))
print(tree_rules)

# === 2. Uji 2 Data Baru ===
uji_baru = pd.DataFrame({
    "Diameter Buah (mm)": [87, 101],
    "Berat Buah (gram)": [295, 330]
})

prediksi = clf.predict(uji_baru)
hasil = le.inverse_transform(prediksi)

print("\n=== Prediksi 2 Data Baru ===")
for i, (d, b) in enumerate(zip(uji_baru['Diameter Buah (mm)'], uji_baru['Berat Buah (gram)'])):
    print(f"Data {i+1}: Diameter={d}, Berat={b} → Prediksi: {hasil[i]}")

# === 3. Visualisasi Decision Boundary ===
print("\n=== Visualisasi Decision Boundary ===")
plot_colors = "ry"
plot_step = 1.0

x_min, x_max = X['Diameter Buah (mm)'].min() - 5, X['Diameter Buah (mm)'].max() + 5
y_min, y_max = X['Berat Buah (gram)'].min() - 10, X['Berat Buah (gram)'].max() + 10

xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                     np.arange(y_min, y_max, plot_step))

grid = np.c_[xx.ravel(), yy.ravel()]
grid_df = pd.DataFrame(grid, columns=['Diameter Buah (mm)', 'Berat Buah (gram)'])
Z = clf.predict(grid_df)
Z = Z.reshape(xx.shape)

plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, cmap=plt.cm.RdYlBu, alpha=0.6)

# Plot titik data training
for label, color in zip([0, 1], plot_colors):
    idx = np.where(y == label)
    plt.scatter(X.iloc[idx]['Diameter Buah (mm)'], X.iloc[idx]['Berat Buah (gram)'],
                c=color, edgecolor='k', label=le.inverse_transform([label])[0])

# Plot data uji baru
plt.scatter(uji_baru['Diameter Buah (mm)'], uji_baru['Berat Buah (gram)'],
            c='black', marker='X', s=100, label='Data Baru')

plt.xlabel("Diameter Buah (mm)")
plt.ylabel("Berat Buah (gram)")
plt.title("Decision Boundary (Diameter vs Berat)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
